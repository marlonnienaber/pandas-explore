```python
import pandas as pd
import numpy as np
import math
from pandas_explore import explore_numerical_attribute 
```

**1) only numerical entiries**  
The function should display that the column contains only numerical values

```python
data = {
    'column': list(range(1, 51)) + [1.0, 3.3337, 76.4657, 34.1]
}

df = pd.DataFrame(data)

explore_numerical_attribute(df["column"])
```

**2) display non-number entries**  
The function should display containen non-numbers and their occurence counts

```python
data = {
    'column': list(range(1, 51)) + [1.0, 3.3337, 76.4657, 34.1] + 
    [np.nan, np.inf, -np.inf, "a", "a", None, math.nan, math.inf, -math.inf]
}

df = pd.DataFrame(data)

explore_numerical_attribute(df["column"])
```

**3) empty column**  
The function should display that the column is empty

```python
data = {
    'column': []
}

df = pd.DataFrame(data)

explore_numerical_attribute(df["column"])
```
